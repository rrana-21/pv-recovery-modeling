```python
#fix date
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(['randid', 'Date']).reset_index(drop=True)
```

# Get target df - getting baseline

```python
records = []

# baseline
window = 7
recovery_threshold = 0.98 

# Loop per plant
for randid, group in df.groupby("randid"):
    group = group.sort_values("Date").reset_index(drop=True)
    
    for i in range(len(group)):
        if group.loc[i, 'weather_onset'] == 1:
            event_date = group.loc[i, 'Date']

            pre_event_data = group[max(0, i - window):i]
            if pre_event_data.empty:
                continue  

            baseline_pr = pre_event_data['PR'].median()

            recovery_found = False
            for j in range(i + 1, len(group)):
                pr_val = group.loc[j, 'PR']
                if not pd.isna(pr_val) and pr_val >= recovery_threshold * baseline_pr:
                    time_to_recovery = (group.loc[j, 'Date'] - event_date).days
                    recovery_found = True
                    break

            if recovery_found:
                records.append({
                    'randid': randid,
                    'event_date': event_date,
                    'time_to_recovery': time_to_recovery,
                    'event_occurred': 1,
                    'baseline_pr': baseline_pr
                })
            else:
                last_date = group['Date'].max()
                time_to_recovery = (last_date - event_date).days
                records.append({
                    'randid': randid,
                    'event_date': event_date,
                    'time_to_recovery': time_to_recovery,
                    'event_occurred': 0,
                    'baseline_pr': baseline_pr
                })

target_df = pd.DataFrame(records)
```

```python
# get date and merge 
df_events = df[df['weather_onset'] == 1].copy()
df_events = df_events.rename(columns={'Date': 'event_date'})

full_df = pd.merge(target_df, df_events, on=['randid', 'event_date'], how='left')
```

```python
full_df = full_df.replace(['None', 'Unknown'], np.nan)
full_df = full_df.fillna(0)

# binary mapping
yes_no_cols = ['active_snow_tickets', 'HurricanePrep', 'HurricanePostInspection', 'storm_active_tickets']
for col in yes_no_cols:
    full_df[col] = full_df[col].map({'Yes': 1, 'No': 0})

region_categoricals = ['NOAAClimRegion', 'TempZone', 'HumidZone', 'bin_PlantSize_kW']

# One-hot encode
categorical_cols = [
    'snow_bin_ticket_minutes', 'snow_affected_assets', 'snow_production_level',
    'hurr_bin_ticket_minutes', 'hurr_affected_assets', 'hurr_production_level',
    'storm_bin_ticket_minutes', 'storm_affected_assets', 'storm_production_level',
    'NOAAClimRegion', 'TempZone', 'HumidZone', 'bin_PlantSize_kW'
]

full_df = pd.get_dummies(full_df, columns=categorical_cols)
```

# Cox Prop Hazard Loss

```python
def init_weights(input_size, hidden_size):
    W1 = np.random.randn(input_size, hidden_size) * 0.01
    b1 = np.zeros((1, hidden_size))
    W2 = np.random.randn(hidden_size, 1) * 0.01
    b2 = np.zeros((1, 1))
    return W1, b1, W2, b2

# Forward pass
def forward(X, W1, b1, W2, b2):
    Z1 = np.dot(X, W1) + b1
    A1 = np.tanh(Z1)
    Z2 = np.dot(A1, W2) + b2
    Z2 = np.clip(Z2, -50, 50) 
    return Z1, A1, Z2

# Cox loss 
def cox_ph_loss(log_hazard, time, event):
    order = np.argsort(time) 
    log_hazard = log_hazard[order]
    event = event[order]

    max_log_hazard = np.max(log_hazard)
    exp_hazard = np.exp(log_hazard - max_log_hazard)
    cum_sum = np.cumsum(exp_hazard)
    log_cumsum = np.log(cum_sum) + max_log_hazard

    observed = log_hazard[event == 1] - log_cumsum[event == 1]
    return -np.mean(observed)

# Training function
def train(X, time, event, hidden_size=32, lr=1e-4, epochs=1000):
    n_samples, n_features = X.shape
    W1, b1, W2, b2 = init_weights(n_features, hidden_size)

    for epoch in range(epochs):
        Z1, A1, Z2 = forward(X, W1, b1, W2, b2)
        log_hazard = Z2.reshape(-1)

        loss = cox_ph_loss(log_hazard, time, event)

        dZ2 = np.ones_like(log_hazard).reshape(-1, 1)
        dW2 = np.dot(A1.T, dZ2)
        db2 = np.sum(dZ2, axis=0, keepdims=True)

        dA1 = np.dot(dZ2, W2.T)
        dZ1 = dA1 * (1 - np.tanh(Z1) ** 2)
        dW1 = np.dot(X.T, dZ1)
        db1 = np.sum(dZ1, axis=0, keepdims=True)

        # Weight updates
        W1 -= lr * dW1
        b1 -= lr * db1
        W2 -= lr * dW2
        b2 -= lr * db2

        if epoch % 100 == 0:
            print(f"Epoch {epoch}, Loss: {loss:.4f}, Max log_hazard: {np.max(log_hazard):.4f}")

    return W1, b1, W2, b2
```

```python
W1, b1, W2, b2 = train(X_scaled, y_time, y_event, hidden_size=32, lr=1e-4, epochs=1000)
```
